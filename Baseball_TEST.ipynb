```python
# Imports

import requests
import pandas 
import numpy as np
```

```python
# Get data

# Source: http://npb.jp/bis/eng/2019/stats/idb1_g.html
g = requests.get("http://npb.jp/bis/eng/2019/stats/idb1_g.html")
t = requests.get("http://npb.jp/bis/eng/2019/stats/idb1_t.html")
```

```python
# Clean data
#DF = data frame (PANDAS)

# Function https://pandas.pydata.org/pandas-docs/version/0.23.4/generated/pandas.read_html.html
dfs = pandas.read_html(g.text, header=1)

# get the first DF from the DFs
df = dfs[0]

# Rename the handedness column
df.rename(columns={ df.columns[0]: "Hand" }, inplace = True)

df.head
```

```python
# Manipulate data

# Replace the weird values with better stuff
df["Hand"] = df["Hand"].replace("*", "L") 
df["Hand"] = df["Hand"].replace("+", "S") 
df["Hand"] = df["Hand"].replace(np.nan, "R") 
df
```

```python
df['random test column']=np.nan
#Create 2 new empty columns 
df.insert(1,'Last',np.nan)
df.insert(2,'First',np.nan)

df
```

```python
# Split Player into 2 new columns 
df[['Last','First']] = df.Player.str.split(",",expand=True)
del df['Player']

```

```python
# Clear white space from first name
df["First"] = df["First"].str.strip()
df
```

```python
# Export data

# https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.to_csv.html
df.to_csv(path_or_buf="/Users/Anzo/Desktop/Tbat.csv")
```
